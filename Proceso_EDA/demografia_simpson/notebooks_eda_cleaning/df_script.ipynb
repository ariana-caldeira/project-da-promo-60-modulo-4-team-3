TENGO QUE UNIR LOS DOS DATAFRAME DE FORMA CORRECTA, NO COINCIDEN ID. LO MÁS INTELIGENTE ES UNIR POR PERSONAJE RESPETANDO LA TABLA DE SCRIPT

```python
df_join = df.copy() #copia tabla script 
```

```python
df_join.head(2)
```

```python
#normalización nombres copia tabla script
df_join["character_name_norm"] = ( 
    df_join["raw_character_text"] 
    .str.lower()
    .str.strip()
)
```

```python
#normalización nombres en characters (por si acaso)
df_characters["normalized_name"] = (
    df_characters["normalized_name"]
    .str.lower()
    .str.strip()
)
```

```python
#left join
df_merged = df_join.merge(
    df_characters,
    how="left",
    left_on="character_name_norm",
    right_on="normalized_name",
    suffixes=("", "_character")
)
```

```python
#✔️ Número de filas igual. Debe devolver True 
df.shape[0] == df_merged.shape[0]
```

```python
#✔️ Porcentaje de líneas que no se han unido
df_merged["gender"].isna().mean() 
```

> 📝 **Documentación:**

El dataset de guiones se utiliza como tabla principal y se enriquece con información del dataset de personajes mediante una unión izquierda basada en el nombre normalizado del personaje.

Esta estrategia permite conservar todas las líneas de guion, incluyendo aquellas que no pueden asociarse a un personaje identificado, garantizando la integridad del análisis narrativo.

```python
df_merged[
    df_merged["gender"].isna()
][["raw_character_text"]].value_counts()
```

> 📝 **Documentación:**

Durante el proceso de integración de datos se observan discrepancias entre los datasets de guiones y personajes. 

Debido al elevado número de personajes episódicos y variantes nominales sin correspondencia clara entre los datasets, se opta por no realizar correcciones manuales exhaustivas. El análisis se centra exclusivamente en personajes con información demográfica identificable, que concentran la mayor parte del diálogo y permiten extraer conclusiones sociales relevantes.

```python
# Se filtran únicamente las líneas de diálogo hablado asociadas a personajes con género identificado, excluyendo acotaciones narrativas y entidades sin correspondencia demográfica.
df_analysis = df_merged[
    (df_merged["speaking_line"] == True) &
    (df_merged["gender"].notna())
]
```

> 📝 **Documentación:**

`df_analysis` es el df final sobre el que vamos a trabajar. Se filtran únicamente las líneas de diálogo hablado asociadas a personajes con género identificado, excluyendo acotaciones narrativas y entidades sin correspondencia demográfica.


```python
df_analysis = df_merged[
    (df_merged["speaking_line"] == "true") &
    (df_merged["gender"].notna())
].copy()
```

```python
df_analysis.head(1)
```

> 📝 Relación clara entre dataframes 


- df: dataset original que contiene todas las líneas del guion de Los Simpsons.

- df_characters: dataset con información demográfica básica de los personajes.

- df_merged: resultado de un left join entre df_scripts y df_characters usando el nombre normalizado del personaje, conservando todas las líneas del guion.

- df_analysis: subconjunto de df_merged que contiene únicamente líneas de diálogo reales pronunciadas por personajes cuyo género ha podido ser identificado.

# Gráficas

```python
# ¿Qué género habla más veces?
df_analysis["gender"].value_counts().plot(kind="bar")
plt.title("Número de líneas de diálogo por género")
plt.xlabel("Género")
plt.ylabel("Número de líneas")
plt.xticks(rotation=90)
plt.show()
```

```python
#¿Quién ocupa más espacio discursivo?
df_analysis.groupby("gender")["word_count"].sum().plot(kind="bar")
plt.title("Cantidad total de palabras por género")
plt.xlabel("Género")
plt.ylabel("Total de palabras")
plt.show()
```

```python
df_analysis["word_count"].dtype
```

```python
df_analysis["word_count"] = pd.to_numeric(
    df_analysis["word_count"],
    errors="coerce"
)
```

```python
df_analysis["word_count"].dtype
```

```python
#¿Quién ocupa más espacio discursivo?
df_analysis.groupby("gender")["word_count"].sum().plot(kind="bar")
plt.title("Cantidad total de palabras por género")
plt.xlabel("Género")
plt.ylabel("Total de palabras")
plt.show()
```

Promedio de palabras por línea según género

```python
df_analysis.groupby("gender")["word_count"].mean().plot(kind="bar")
plt.title("Promedio de palabras por línea según género")
plt.xlabel("Género")
plt.ylabel("Palabras por línea")
plt.show()
```

¿Quién concentra el discurso en la serie?

```python
(
    df_analysis
    .groupby("raw_character_text")["word_count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .plot(kind="barh")
)

plt.title("Top 10 personajes por cantidad de palabras")
plt.xlabel("Total de palabras")
plt.ylabel("Personaje")
plt.show()
```

¿Quiénes son los personajes con más voz y qué género tienen?

```python
top10 = (
    df_analysis
    .groupby("raw_character_text")["word_count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

df_analysis[df_analysis["raw_character_text"].isin(top10)] \
    .groupby("gender")["word_count"] \
    .sum() \
    .plot(kind="bar")

plt.title("Distribución por género dentro del top 10 personajes")
plt.xlabel("Género")
plt.ylabel("Total de palabras")
plt.show()
```

Género dentro del top 10 personajes

```python
top10 = (
    df_analysis
    .groupby("raw_character_text")["word_count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

df_analysis[df_analysis["raw_character_text"].isin(top10)] \
    .groupby("gender")["word_count"] \
    .sum() \
    .plot(kind="bar")

plt.title("Distribución por género dentro del top 10 personajes")
plt.xlabel("Género")
plt.ylabel("Total de palabras")
plt.show()
```
