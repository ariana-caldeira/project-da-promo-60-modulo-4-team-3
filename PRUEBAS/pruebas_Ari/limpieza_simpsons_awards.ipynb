## **LIMPIEZA DE COLUMNAS Y UNION DE DATOS**

#### **THE SIMPSONS - AWARDS & RATINGS**

```python
import pandas as pd
import numpy as np
from src import eda_soporte as eda_sp

pd.set_option('display.max_columns', None)
```

```python
df_award = eda_sp.open_csv("./datasets/award.csv")
df_award.head(1)
```

```python
# ELIMINAR 3 columnas - 98% nulos y tenemos esta info en otros datasets

df_award = df_award.drop(columns=['season', 'song', 'role'])
```

```python
df_character_award = eda_sp.open_csv("./datasets/character_award.csv")
df_character_award.head(1)
```

```python
df_award = df_award.merge(df_character_award, on='award_id', how='left')
```

```python
df_award.shape
```

```python
df_credit = eda_sp.open_csv("./datasets/credit.csv")
df_credit.head(1)
```

```python
#Limpiar duplicados 
df_credit = df_credit.drop_duplicates()
```

```python
df_credit.shape
```

```python
# ELIMINAR 'credited' - la info no es √∫til

df_credit = df_credit.drop(columns=['credited', 'category'])
```

```python
df_award = df_award.merge(df_credit[['episode_id', 'person', 'role']],
            on=['episode_id', 'person'],
            how='left')
```

```python
df_award.shape
```

```python
df_episode = eda_sp.open_csv("./datasets/episode.csv")
df_episode.head(1)
```

```python
# Solo me interesan 2 columnas de este dataset para hacer de puente con el df_ratings

df_episode = df_episode.drop(columns=['number_in_series', 'title',
       'summary', 'air_date', 'episode_image', 'rating', 'votes'])
```

```python
df_award = df_award.merge(df_episode, on='episode_id', how='left')
```

```python
df_award.columns
```

```python
df_award.shape
```

```python
# Organizar el orden de las columnas

cols = ['award_id', 'organization', 'year', 'award_category', 'award', 'result', 
        'person', 'role', 'character',
        'episode_id', 'season', 'episode']

df_award = df_award[cols]
```

```python
# Verificar si no hay duplicados y el orden de las columnas
eda_sp.eda_1(df_award)
```

```python
df_award = df_award.drop_duplicates()
```

```python
df_award.duplicated().sum()
```

```python
df_award.to_csv("./final_data/simpsons_awards.csv", index=False)
```
