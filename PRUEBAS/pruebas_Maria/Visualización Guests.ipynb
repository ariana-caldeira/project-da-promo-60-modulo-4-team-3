## **VISUALIZACIÓN GUESTS**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('simpsons-guests.csv', sep='|')
df.head()
```

```python
roles_himself_herself = df[
    df['role'].str.lower().str.contains('himself', na=False) |
    df['role'].str.lower().str.contains('herself', na=False)]
```

```python
roles_himself_herself.sample(5)
```

Lo primero que hice fue filtrar el dataset para ver cuáles invitados se habían interpretadoa a si mismos y no habían doblado a otros personajes

```python
roles_of_self = df[df['role'].isin(['Himself', 'Herself'])]

roles_by_season = roles_of_self.groupby('season').size().reset_index(name='count_himself_herself')

roles_by_season
```

Luego los agrupé por temporada para saber cuántos invitados había tenido cada temporada

```python
roles_by_pc = roles_of_self.groupby('production_code').size().reset_index(name='count_himself_herself') 
roles_by_pc
```

```python
roles_by_pc_ascendente = roles_of_self.groupby('production_code').size().reset_index(name='count_himself_herself').sort_values(by='count_himself_herself', ascending=True)

roles_by_pc_ascendente.tail(10)
```
