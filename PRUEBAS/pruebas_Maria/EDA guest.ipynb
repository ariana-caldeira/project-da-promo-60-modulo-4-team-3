```python
# Contar el número de episodios por cada temporada
conteo_episodios = df.groupby('season')['episode_title'].count().reset_index(name='Total_Episodios')

print("\nConteo de episodios por temporada:")
print(conteo_episodios)
```

```python

# Ordenar por la columna 'Total_Episodios' de forma ascendente
conteo_ascendente_episodios = conteo_episodios.sort_values(by='Total_Episodios', ascending=True)

print("Temporadas ordenadas por el Conteo de Episodios (Ascendente):")
print(conteo_ascendente_episodios)
```
