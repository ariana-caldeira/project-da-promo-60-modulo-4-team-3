Vemos que en la columna sector, hay algunos invitados que tienen más de una profesión. Con el fin de que no nos de problemas a la hora de trabajarlo en Tableau, vamos a priorizar la primera profesión, y vamos a eliminar la segunda.

```python
#Separamos en dos columnas distintas las distintas profesiones en caso de que existan
df_final[['sector_1', 'sector_2']] = df_final['sector'].str.split('/', n=1, expand=True)

df_final['sector_1'] = df_final['sector_1'].str.strip()
df_final['sector_2'] = df_final['sector_2'].str.strip()

print("¡Archivo procesado! Se han creado las columnas 'sector_1' y 'sector_2'.")
```
