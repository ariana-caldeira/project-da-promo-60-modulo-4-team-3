TENGO QUE UNIR LOS DOS DATAFRAME DE FORMA CORRECTA, NO COINCIDEN ID. LO MÁS INTELIGENTE ES UNIR POR PERSONAJE RESPETANDO LA TABLA DE SCRIPT

```python
df_join = df.copy() #copia tabla script 
```

```python
df_join.head(2)
```

```python
#normalización nombres copia tabla script
df_join["character_name_norm"] = ( 
    df_join["raw_character_text"] 
    .str.lower()
    .str.strip()
)
```

```python
#normalización nombres en characters (por si acaso)
df_characters["normalized_name"] = (
    df_characters["normalized_name"]
    .str.lower()
    .str.strip()
)
```

```python
#left join
df_merged = df_join.merge(
    df_characters,
    how="left",
    left_on="character_name_norm",
    right_on="normalized_name",
    suffixes=("", "_character")
)
```

```python
#✔️ Número de filas igual. Debe devolver True 
df.shape[0] == df_merged.shape[0]
```

```python
#✔️ Porcentaje de líneas que no se han unido
df_merged["gender"].isna().mean() 
```

> 📝 **Documentación:**

El dataset de guiones se utiliza como tabla principal y se enriquece con información del dataset de personajes mediante una unión izquierda basada en el nombre normalizado del personaje.

Esta estrategia permite conservar todas las líneas de guion, incluyendo aquellas que no pueden asociarse a un personaje identificado, garantizando la integridad del análisis narrativo.

```python
df_merged[
    df_merged["gender"].isna()
][["raw_character_text"]].value_counts()
```

> 📝 **Documentación:**

Durante el proceso de integración de datos se observan discrepancias entre los datasets de guiones y personajes. 

Debido al elevado número de personajes episódicos y variantes nominales sin correspondencia clara entre los datasets, se opta por no realizar correcciones manuales exhaustivas. El análisis se centra exclusivamente en personajes con información demográfica identificable, que concentran la mayor parte del diálogo y permiten extraer conclusiones sociales relevantes.

```python
# Se filtran únicamente las líneas de diálogo hablado asociadas a personajes con género identificado, excluyendo acotaciones narrativas y entidades sin correspondencia demográfica.
df_analysis = df_merged[
    (df_merged["speaking_line"] == True) &
    (df_merged["gender"].notna())
]
```
